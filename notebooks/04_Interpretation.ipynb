## 1. Data


```python
from pathlib import Path
import pandas as pd

ROOT = Path.cwd().resolve().parent if Path.cwd().name == "notebooks" else Path.cwd().resolve()

TAB_DIR = ROOT / "outputs" / "tables"
SUM_DIR = ROOT / "outputs" / "summary"
SUM_DIR.mkdir(parents=True, exist_ok=True)

descriptive_statistics = pd.read_csv(TAB_DIR / "descriptive_statistics.csv")
anova_results = pd.read_csv(TAB_DIR / "anova_results.csv")
tukey_results = pd.read_csv(TAB_DIR / "tukey_hsd_results.csv")
inference_summary = pd.read_csv(TAB_DIR / "03_inference_summary.csv")

print("Descriptive statistics:")
display(descriptive_statistics)

print("ANOVA results:")
display(anova_results)

print("Tukey HSD results:")
display(tukey_results)

print("Inference summary:")
display(inference_summary)
```

## 2. Final Interpretation


The cleaned analysis dataset contained **12,853** students.

The group with the highest mean BMI percentile was **Very overweight**, with mean BMIPCT = **92.30**.

The group with the lowest mean BMI percentile was **Slightly underweight**, with mean BMIPCT = **36.53**.

The one-way ANOVA result was **F = 1556.6455** with **p-value < 0.001**. At alpha = 0.05, the decision is to **Reject H0**.

Because the ANOVA result is statistically significant, Tukey HSD post-hoc comparisons were used. In this analysis, **10 of 10** pairwise comparisons were statistically significant.

Overall, students' perceived weight category is associated with their actual BMI percentile. Students who perceived themselves as overweight generally had higher BMI percentiles than students who perceived themselves as underweight or about the right weight.

This result shows an **association**, not causation, because the YRBS data are observational survey data.


## 3. Limitations


- This is observational survey data, so the result shows association, not causation.
- BMI percentile does not fully capture health status.
- Self-reported survey answers may contain reporting bias.
- Other variables, such as age, sex, physical activity, and nutrition, were not included in this one-way ANOVA.
